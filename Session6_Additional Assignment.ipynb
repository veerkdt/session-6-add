1. Read the dataset using pandas.

```python
import pandas as pd
```

```python
ratings=pd.read_csv("ratings.csv")
```

```python
tags=pd.read_csv("tags.csv")
```

```python
movies=pd.read_csv("movies.csv")
```

4. Print index, columns of the DataFrame.

```python
print("****************************\n index, columns of the ratings dataframe \n ****************************")
ratings.info()
print("****************************\n index, columns of the tags dataframe \n ****************************")
tags.info()
print("****************************\n index, columns of the movies dataframe \n ****************************")
movies.info()
```

5. Calculate descriptive statistics for the 'ratings' column of the ratings DataFrame. Verify using describe().

```python
print("individual calculations over rating column")
print("countc:",ratings.rating.count())
print("Mean:",ratings.rating.mean())
print("Std:",ratings.rating.std())
print("Min:",ratings.rating.min())
print("Max:",ratings.rating.max())
print("using decribe function")
ratings.rating.describe()
```

6. Filter out ratings with rating > 5

```python
ratings[ratings["rating"]>=5]
```

7. Find how many null values, missing values are present. Deal with them. Print out how many rows have been modified.

```python
print("***Count NaN, missing values in each column of a Ratings DataFrame***")
print("Nan in each columns" , ratings.isnull().sum(), sep='\n')
print("***Count NaN, missing values in each column of a Tags DataFrame***")
print("Nan in each columns" , tags.isnull().sum(), sep='\n')
print("***Count NaN, missing values in each column of a Movies DataFrame***")
print("Nan in each columns" , movies.isnull().sum(), sep='\n')

print("Filling Null or missing values with Santosh")
Dataset=tags.fillna("Santosh")
print("Modified row count is\n",Dataset[Dataset["tag"]=="Santosh"].count())
```

8. Filter out movies from the movies DataFrame that are of type 'Animation'.

```python
df = movies[movies['genres']!='Animation']

print(df)
```

9. Find the average rating of movies.

```python
print("Average rating of the movies:",ratings.rating.mean())
```

10. Perform an inner join of movies and tags based on movieId.

```python
movies.join(tags,on='movieId',how='inner',lsuffix='Mov')
```

11. Print out the 5 movies that belong to the Comedy genre and have rating greater than 4.

```python
Df1 = movies.join(ratings,on='movieId',how='inner',lsuffix='Mov')
df2 = Df1[(Df1['genres']=='Comedy') & (Df1['rating']>4.0)]
df2.head(5)
```

14. Select rows based on timestamps later than 2015-02-01.

```python
import datetime

df=ratings
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date
D=datetime.date(2015,2,1)
res=df[df['timestamp'] > D]
res
```
